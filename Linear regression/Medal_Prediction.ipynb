```python
#Importing the libraries
import pandas as pd
```

```python
#Importing the dataset - credits : https://www.kaggle.com/the-guardian/olympic-games
teams = pd.read_csv("teams.csv")
```

```python
#Data Preprocessing to check if the data is clean
teams
```

```python
teams = teams[["team", "country", "year", "athletes", "age", "prev_medals", "medals"]]
```

```python
teams
```

```python
import numpy as np

# Select only numeric columns
numeric_teams = teams.select_dtypes(include=[np.number])

# Calculate correlation
correlation = numeric_teams.corr()["medals"]

correlation
```

```python
teams=teams.dropna()
teams
```

```python
#Splitting the dataset into the Training set and Test set
train = teams[teams["year"] < 2012].copy()
test = teams[teams["year"] >= 2012].copy()  
```

```python
#To check the size of the training set
train.shape
```

```python
#To check the size of the test set
test.shape
```

```python
#Importing the libraries from sci-kit learn
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors], train["medals"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"] = predictions
test
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0    
test
```

```python
test["predictions"] = test["predictions"].round().astype(float)
test
```

```python
from sklearn.metrics import mean_absolute_error
error  = mean_absolute_error(test["medals"], test["predictions"])
error
```
